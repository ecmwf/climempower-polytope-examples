```python
import requests

url = "https://polytope-dss.ecmwf.int/api/v1/dataset/sis-hydrology-variables-derived-projections/extract/?download_file=test.zip"
headers = {
    "accept": "application/json",
    "Content-Type": "application/json",
}
data = {
    "request": {
        "product_type": "climate_impact_indicators",
        "variable": ["river_discharge"],
        "variable_type": "absolute_values",
        "time_aggregation": "annual_mean",
        "experiment": ["historical"],
        "hydrological_model": ["e_hypegrid"],
        "rcm": "csc_remo2009",
        "gcm": "mpi_esm_lr",
        "ensemble_member": ["r1i1p1"],
        "period": ["1971_2000"],
    },
    "feature": {
        "type": "boundingbox",
        "coordinates": [
            [49.5, 5.5],
            [55.5, 15.5],
        ],
    },
}

response = requests.post(url, headers=headers, json=data)

# Print the response
print(response.status_code)
print(response.json())
```
