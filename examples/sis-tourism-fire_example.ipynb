```python
import requests

url = "https://polytope-dss.ecmwf.int/api/v1/dataset/sis-tourism-fire-danger-indicators/extract/?download_file=test.zip"
headers = {
    "accept": "application/json",
    "Content-Type": "application/json",
}
data = {
    "request": {
        "time_aggregation": "seasonal_indicators",
        "product_type": "single_model",
        "variable": ["seasonal_fire_weather_index"],
        "gcm_model": ["ec_earth"],
        "experiment": "historical",
        "period": ["2001_2005"],
        "version": "v1_0",
    },
    "feature": {
        "type": "boundingbox",
        "coordinates": [[0.0,0.0], [1.0, 1.0]],
    },
}

response = requests.post(url, headers=headers, json=data)

# Print the response
print(response.status_code)
print(response.json())
```

```python
from covjsonkit.api import Covjsonkit
decoder = Covjsonkit().decode(response.json())

ds = decoder.to_xarray()
ds
```
