```python
import requests

url = "https://polytope-dss.ecmwf.int/api/v1/dataset/sis-heat-and-cold-spells/extract/?download_file=test.zip"
headers = {
    "accept": "application/json",
    "Content-Type": "application/json",
}
data = {
    "request": {
        "variable": ["cold_spell_days"],
        "definition": "country_related",
        "experiment": ["rcp8_5"],
        "ensemble_statistic": ["ensemble_members_average"]
    },
    "feature": {
        "type": "boundingbox",
        "coordinates": [[49.5, 5.5], [50.5, 6.5]],
    },
}

response = requests.post(url, headers=headers, json=data)

# Print the response
print(response.status_code)
print(response.json())
```

```python
from covjsonkit.api import Covjsonkit
decoder = Covjsonkit().decode(response.json())

ds = decoder.to_xarray()
ds
```

```python
import requests

url = "https://polytope-dss.ecmwf.int/api/v1/dataset/sis-heat-and-cold-spells/extract/?download_file=test.zip"
headers = {
    "accept": "application/json",
    "Content-Type": "application/json",
}
data = {
    "request": {
        "variable": ["heat_wave_days"],
        "definition": "health_related",
        "experiment": ["rcp4_5"],
        "ensemble_statistic": ["ensemble_members_average"],
    },
    "feature": {
        "type": "polygon",
        "coordinates": [
            [
                49.5,
                5.5
            ],
            [
                50.0,
                5.5
            ],
            [
                50.0,
                6.0
            ],
            [
                49.5,
                6.0
            ],
            [
                49.5,
                5.5
            ],
        ],
    },
}

response = requests.post(url, headers=headers, json=data)

# Print the response
print(response.status_code)
print(response.json())
```

```python
from covjsonkit.api import Covjsonkit
decoder = Covjsonkit().decode(response.json())

ds = decoder.to_xarray()
ds
```
